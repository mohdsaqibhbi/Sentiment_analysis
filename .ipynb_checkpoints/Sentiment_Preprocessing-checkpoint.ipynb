## Cleaning Subset train data

#### Stopwords

```python
nltk.download('stopwords')
stopwords = list(nltk.corpus.stopwords.words('english'))
print(sorted(stopwords))
```

### Cleaning text

#### clean object

```python
clean = CleanText(contractions, stopwords)
```

#### Data before cleaning

```python
train_df.head()
```

#### Data after cleaning

```python
train_df["comment_text"] = train_df.comment_text.apply(clean.get_alphabetic)
train_df["comment_text"] = train_df.comment_text.apply(clean.expand_contracted)
train_df["comment_text"] = train_df.comment_text.apply(clean.remove_stopwords)
train_df["comment_text"] = train_df.comment_text.apply(clean.remove_apostrophe)
train_df.head()
```

```python
def find_word_frequency(row):
    global word_frequency
    word_frequency+= Counter(row.lower().split())
```

## Read Glove vector

```python
# Read glove vector file
def read_glove_vecs(glove_file):
    with open(glove_file, 'r') as f:
        words = set()
        word_to_vec_map = {}
        for line in f:
            line = line.strip().split()
            curr_word = line[0]
            words.add(curr_word)
            word_to_vec_map[curr_word] = np.array(line[1:], dtype=np.float64)
        
        i = 1
        words_to_index = {}
        index_to_words = {}
        for w in sorted(words):
            words_to_index[w] = i
            index_to_words[i] = w
            i = i + 1
    return words_to_index, index_to_words, word_to_vec_map
```

```python
words_to_index, index_to_words, word_to_vec_map = read_glove_vecs('data/glove.6B.50d.txt')
```

### Words not present in the embeddings vector(vocab)

#### Vocab

```python
vocab = list(words_to_index.keys())
```

#### Words in the training data

```python
word_list = list(word_frequency.keys())
```

#### Words not in the vocab

```python
def not_in_vocab(word):
    
    if word not in vocab:
        return (word, word_frequency[word])
```

```python
non_vocab = list(map(not_in_vocab, word_list))
```

```python
non_vocab = list(filter(lambda a: a != None, non_vocab))
len(non_vocab)
```
